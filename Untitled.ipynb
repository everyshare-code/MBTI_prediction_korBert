```python
import glob,os
# 특정 디렉토리의 모든 txt 파일 목록을 가져옵니다.
file_list = glob.glob('./korBert/analysis/mbti_top_words_200/*.txt')

result_dict = {}

print(file_list[0])
# 목록을 순회하며 각 파일을 읽습니다.
for file_path in file_list:
    # 파일 경로에서 파일명만 추출 (예: 'enfj_top_words.txt')
    file_name = os.path.basename(file_path)
    # 파일명에서 키 추출 (예: 'enfj')
    key = file_name.split('_')[0]
    with open(file_path, 'r', encoding='utf-8') as file:
        # 파일의 모든 줄을 읽고 처음 20줄만 선택
        lines = [line.strip() for line in file.readlines()[:30]]
        result_dict[key] = lines
    # 수정된 내용을 딕셔너리에 저장
    
```

```python
mbti_words = {
    'E': [],
    'I': [],
    'S': [],
    'N': [],
    'T': [],
    'F': [],
    'J': [],
    'P': []
}

# # Original MBTI words mapping (partial for demonstration)
# original_mapping = {
#     'enfj': ['사람', '다른', '선도', '수', '이', '자신', '감정', '일', '생각', '유형', '데', '편이', '위해', '것', '때', '때문', '능력', '타인', '자기', '성격', '상대방', '더', '가장', '친구', '대해', '삶', '대화', '성향', '말', '적극'],
#     # Add other MBTI types here...
# }

# Iterate over each MBTI type and its words
for mbti_type, words in result_dict.items():
    # Determine dichotomies
    dichotomies = list(mbti_type.lower())
    for dichotomy in dichotomies:
        # Skip if the current MBTI type dichotomy is not in the target groups
        if dichotomy.upper() not in mbti_words:
            continue
        
        # Add words to the corresponding group if not already present
        for word in words:
            if word not in mbti_words[dichotomy.upper()]:
                mbti_words[dichotomy.upper()].append(word)

# For demonstration, let's print the first 5 words for each dichotomy
for dichotomy, words in mbti_words.items():
    print(f"{dichotomy}: {words[:5]}")
```

```python
import pandas as pd
from transformers import AutoTokenizer
def split_and_expand_descriptions(df, max_length=64):
    tokenizer = AutoTokenizer.from_pretrained('monologg/koelectra-base-v3-generator')
    expanded_rows = []  # 확장된 행들을 저장할 리스트

    for _, row in df.iterrows():
        mbti = row['mbti']
        text = row['description']

        tokens = tokenizer.tokenize(text)
        # 토큰의 총 길이가 최대 길이 이하이면, 해당 행을 그대로 추가
        if len(tokens) <= max_length:
            expanded_rows.append({'mbti': mbti, 'description': text})
        else:
            # 최대 길이를 초과하면, 쪼개어 추가
            start = 0
            while start < len(tokens):
                split_tokens = tokens[start:start+max_length]
                split_text = tokenizer.convert_tokens_to_string(split_tokens)
                expanded_rows.append({'mbti': mbti, 'description': split_text})
                start += max_length

    # 새로운 데이터프레임 생성
    expanded_df = pd.DataFrame(expanded_rows)
    return expanded_df
```
